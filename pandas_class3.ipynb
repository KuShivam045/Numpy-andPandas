```python
df_sales = pd.read_csv(r'sales_data_final.csv')
df_sales
```

```python
def profit_flag(a):
    if a >= 0:
        return "positive"
    else:
        return "negative"
```

```python
df_sales["flag_profit"] = df_sales["profit"].apply(profit_flag)

df_sales.head()
```

```python
def quantity_flag(a):
    if a <= 2:
        return 'low'
    elif a > 2 and a <= 4:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
df_sales.head(20)
```
