# 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset

```python
def calorie_check(b):
    if b >= 1800:
        return 1
    else:
        return 0
```

```python
df['HealthyDay'] = df['Calories'].apply(calorie_check)
```

```python
df[['Id', 'HealthyDay']]
```

```python
healthy_month = df.groupby('Id')[['HealthyDay', 'Calories']].sum()
```

```python
healthy_month.sort_values(by = "Calories", ascending = False).head()
```

```python
healthy_month["LoggedDays"] = df.groupby("Id")['ActivityDate'].count().sort_values(ascending = False)
```

```python
healthy_month['AvgCalories'] = healthy_month['Calories']/healthy_month['LoggedDays']
```

```python
healthy_month["AvgHealthIndex"] = healthy_month['HealthyDay']/healthy_month['LoggedDays']
```

```python
healthy_month.sort_values(by = ['AvgCalories', "HealthyDay"], ascending = False).head()
```

```python
def Healthstatus(a,b):
    if a >= 1800 or b > 0.8:
        return 'Healthy'
    else:
        return 'Unhealthy'
```

```python
healthy_month['HealthStatus']= healthy_month.apply(lambda healthy_month: Healthstatus(healthy_month['AvgCalories'],healthy_month['AvgHealthIndex']),axis= 1)
healthy_month
```

```python
healty_count = healthy_month[healthy_month["HealthStatus"] == "Healthy"]
```

```python
len(healty_count.index)
```
